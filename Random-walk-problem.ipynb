<h4>Example-2: Random walk problem</h4> 

**Rules:**

- If the direction is 'up', it increments the y-coordinate by 1.
- If the direction is 'down', it decrements the y-coordinate by 1.
- If the direction is 'left', it decrements the x-coordinate by 1.
- If the direction is 'right', it increments the x-coordinate by 1.

```python
import random
import matplotlib.pyplot as plt

def random_walk(n):
    """Simulates a random walk of 'n' steps."""
    x = 0  # Starting position
    y = 0
    x_values = [x]
    y_values = [y]

    # Perform 'n' steps of the random walk
    for _ in range(n):
        direction = random.choice(['up', 'down', 'left', 'right'])
        if direction == 'up':
            y += 1
        elif direction == 'down':
            y -= 1
        elif direction == 'left':
            x -= 1
        elif direction == 'right':
            x += 1
        x_values.append(x)
        y_values.append(y)

    return x_values, y_values

# Set the number of steps for the random walk
num_steps = 1000

# Simulate the random walk
x, y = random_walk(num_steps)

# Plot the random walk
plt.plot(x, y)
plt.title('Random Walk')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

<h4>Example-3: Will our drunk friend make it to the bathroom?</h4> 

**Rules:**
- Let's consider a scenario where our drunk friend is initially located at position $(0, 0)$ on a grid and needs to reach the bathroom located at position $(x_{\text{bathroom}}, y_\text{bathroom})$. 
- Our friend can only move one step at a time in either the vertical or horizontal direction.

**Logic:** 
- The logic to determine if our drunk friend will make it to the bathroom is based on the Manhattan distance between the starting position and the bathroom. 
- The Manhattan distance is the sum of the absolute differences in the x and y coordinates. 
- If the Manhattan distance is less than or equal to the number of steps our friend is willing to take, then they can reach the bathroom. Otherwise, they won't be able to reach it.

```python
import random

def will_reach_bathroom(x_bathroom, y_bathroom, num_steps):
    x = 0  # Starting x-coordinate
    y = 0  # Starting y-coordinate

    # Calculate the Manhattan distance
    distance = abs(x_bathroom - x) + abs(y_bathroom - y)

    # Check if the distance is less than or equal to the number of steps
    if distance <= num_steps:
        return True
    else:
        return False

# Set the coordinates of the bathroom and the number of steps
x_bathroom = 4
y_bathroom = 7
num_steps = 6

# Check if our drunk friend will make it to the bathroom
result = will_reach_bathroom(x_bathroom, y_bathroom, num_steps)

# Print the result
if result:
    print("Our drunk friend will make it to the bathroom!")
else:
    print("Our drunk friend won't make it to the bathroom!")
```

```python
import random
import matplotlib.pyplot as plt

def simulate_random_walk(x_bathroom, y_bathroom, num_steps):
    x = 0  # Starting x-coordinate
    y = 0  # Starting y-coordinate
    x_values = [x]  # List to store x-coordinates of each step
    y_values = [y]  # List to store y-coordinates of each step

    for step in range(num_steps):
        direction = random.choice(['up', 'down', 'left', 'right'])

        if direction == 'up':
            y += 1
        elif direction == 'down':
            y -= 1
        elif direction == 'left':
            x -= 1
        elif direction == 'right':
            x += 1

        x_values.append(x)
        y_values.append(y)

        if (x, y) == (x_bathroom, y_bathroom):
            return x_values, y_values

    return x_values, y_values

# Set the coordinates of the bathroom and the number of steps
x_bathroom = 3
y_bathroom = 2
num_steps = 20

# Simulate the random walk to the bathroom
x, y = simulate_random_walk(x_bathroom, y_bathroom, num_steps)

# Print the result
if (x[-1], y[-1]) == (x_bathroom, y_bathroom):
    print("Your friend will reach the bathroom within {} steps!".format(num_steps))
else:
    print("Your friend won't reach the bathroom within {} steps.".format(num_steps))

# Plot the path with solid red line
plt.plot(x, y, '-o', color='red')
plt.scatter(x_bathroom, y_bathroom, color='blue', label='Bathroom')
plt.title("Path to the Bathroom ({} steps)".format(num_steps))
plt.xlabel("X-coordinate")
plt.ylabel("Y-coordinate")
plt.legend()
plt.grid(True)
plt.show()
```

To simulate the random movement of your drunk friend in any direction and visualize the path to the bathroom, you can use the following modified Python code:

```python
import random
import math
import matplotlib.pyplot as plt

def simulate_random_walk(x_start, y_start, x_bathroom, y_bathroom, num_steps):
    x = x_start  # Starting x-coordinate
    y = y_start  # Starting y-coordinate
    x_values = [x]  # List to store x-coordinates of each step
    y_values = [y]  # List to store y-coordinates of each step

    for _ in range(num_steps):
        angle = random.uniform(0, 2 * 3.14159)  # Random angle in radians
        distance = random.uniform(0, 1)  # Random distance in any direction

        x += distance * (x_bathroom - x) * abs(math.cos(angle))
        y += distance * (y_bathroom - y) * abs(math.sin(angle))

        x_values.append(x)
        y_values.append(y)

        if (x, y) == (x_bathroom, y_bathroom):
            break

    return x_values, y_values

# Set the coordinates of the starting point and the bathroom
x_start = 0
y_start = 0
x_bathroom = 3
y_bathroom = 2

# Set the number of steps for the random walk
num_steps = 20

# Simulate the random walk to the bathroom
x, y = simulate_random_walk(x_start, y_start, x_bathroom, y_bathroom, num_steps)

# Plot the path with solid red line
plt.plot(x, y, '-o', color='red', linewidth=2)
plt.scatter(x_start, y_start, color='green', label='Starting Point')
plt.scatter(x_bathroom, y_bathroom, color='blue', label='Bathroom')
plt.title("Path to the Bathroom")
plt.xlabel("X-coordinate")
plt.ylabel("Y-coordinate")
plt.legend()
plt.grid(True)
plt.show()
```
