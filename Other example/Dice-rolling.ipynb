**Example-1** Simulating rolling a fair six-sided die and calculating the average value after a large number of rolls.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#  Python code that simulates rolling a fair six-sided die and calculates the average value after a large number of rolls:

import random

num_rolls = 1000000  # Number of rolls
rolls = []  # List to store rolled values
sum_values = 0  # Variable to store the sum of rolled values

# Simulate rolling the die num_rolls times
for _ in range(num_rolls):
    roll = random.randint(1, 6)  # Randomly generate a number between 1 and 6
    rolls.append(roll)
    sum_values += roll

average = sum_values / num_rolls  # Calculate the average value

print(f"The average value after {num_rolls} rolls is: {average}")
```

In the loop `for _ in range(num_rolls)`:, the underscore `_` is commonly used as a variable name when you don't intend to use the value of the iterator variable.

```python
# Plotting the histogram
plt.hist(rolls, bins=6, range=(1, 7), edgecolor='red', rwidth=0.8)
plt.xlabel('Rolled Value')
plt.ylabel('Frequency')
plt.title('Distribution of Rolled Values')
plt.grid(True)
plt.show()
```

```python
# Count the frequencies of each rolled value
value_counts = [rolls.count(i) for i in range(1, 7)]

# Plotting the pie chart
plt.pie(value_counts, labels=range(1, 7), autopct='%1.1f%%')
plt.title('Relative Frequencies of Rolled Values')
plt.axis('equal')  # Equal aspect ratio ensures a circular pie
plt.show()
```
